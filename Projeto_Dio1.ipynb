### 1. Importação dos pacotes e Preparação dos Dados 

```python
# Importando o pacote pandas
import pandas as pd
```

```python
# Importando o matplotlib
import matplotlib.pyplot as plt
```

```python
# Importando o dataframe a ser trabalhado
df = pd.read_table('diarias.csv', sep = ';')
```

```python
df
```

### 2.Tratamento e limpeza dos dados

```python
# Verificando as colunas
df.columns.values
```

```python
# Selecionando as colunas
df2 = ['cargo_beneficiario', 'valor', 'dt_concessao']
```

```python
# Filtrando as colunas selecionadas para o dataframe trabalhado
df = df.filter(items=df2)
```

```python
# Removendo valores nulos da tabela
df = df.dropna()
```

```python
# Listando a quantidade de Cargos
df.groupby('cargo_beneficiario').count()
```

```python
# Mostrando os cargos que tiveram gastos maiores que 100 e somando os valores dos gastos de cada cargo
df[df.valor > 100].groupby('cargo_beneficiario').sum()
```

```python
df3 = df[df.valor > 100].groupby('cargo_beneficiario').sum()
```

```python
df3
```

```python
cargo_selecionado = ['cargo_beneficiario']
```

```python
cargo = df3.filter(items=cargo_selecionado)
```

```python
cargo
```

```python
# Reiniciando o index da coluna e separando em duas colunas
df3 = df3.reset_index()
```

```python
valor_selecionado = ['valor']
```

```python
valores = df3.filter(items=valor_selecionado)
```

```python
valores
```

### 3. Visualização dos Dados


```python
plt.figure(figsize=(15,7))
plt.barh(df3.cargo_beneficiario, df3.valor)
plt.ylabel('Cargos_Beneficiarios')
plt.xlabel('Valores')
plt.title('Gastos com diárias')
plt.show()
```
